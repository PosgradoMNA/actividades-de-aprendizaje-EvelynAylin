#### Sets (5:12)
Set { }. Son un tipo de colecciones, por lo que es posible introducir distintos tipos de datos. Se diferencian de las listas y tuplas (que tambin son colecciones) porque no son ordenados, lo que significa que no graba la posicin del elemento. Asimismo, se caracterizan porque tienen elementos nicos, es decir uno mismo no se puede repetir en estos conjuntos (no se pueden duplicar) y se generan a partir de llaves 

```python
#Creacin
set1={'pop','rock','opera','pop','soul','pop'}
print(set1) #No se muestran los elementos repetidos
print(type(set1))
#Operaciones
lista=["hola",1,2,"hola"]
print(lista)
setDeLista=set(lista) #type-casting, convertir una lista a set
print(setDeLista) 
setDeLista.remove("hola") #remover elementos de un set
print(setDeLista)
print("hola" in setDeLista) #Verificar si existe determinado elemento en el set con in
print(type("hola" in setDeLista)) #El resultado es un valor bool

print("")
print("El smbolo & encuentra la unin entre dos sets")
# Al usar & para concatenar dos sets, solo mantendr los elementos que ambos tengan en comn
setA={'yo','s','soy'}
setB={'yo','no','soy'}
print("Set A:",setA, "\nSet B:",setB)
setC=setA&setB
print("setA&setB:",setC) #Solo se mantiene 'yo', 'soy' porque son los nicos elementos compartidos

print("")
print("union() concatena todos los elementos de los sets en uno nuevo")
setD=setA.union(setB)
print(setD) #Junta todos lo elementos pero no los repite
Set2={'yo','soy'}
setE=setA.union(Set2) #Set2 no aporta informacin nueva a set E, por lo que se encapsula
print(setE) 
print(".issubset() nos permite verificar si un set est encapsulado en el set argumento")
print(Set2.issubset(setA)) #La respuesta se da a travs de un valor bool

#------Sets (External resource) LAB------
print(setA.issuperset(Set2)) #Busca si set A es un super set de set2 (encapsula al segundo)
print(setA.difference(setB)) #Encuentra lo que diferencia el setA del B
print(setA.intersection(setB)) #)Encuentra la interseccin o que es igual en el setA respecto el B

#-------Quiz on Sets-------
#Cast the following list to a set:
['A','B','C','A','B','C']
print(set(['A','B','C','A','B','C']))
#Add the string 'D' to the set S.
S={'A','B','C'}
print(S)
S.add('D')
print(S)
#Find the intersection of set A and B.
A={1,2,3,4,5}
B={1,3,9, 12}
print(A.intersection(B))
#What is the result of the following lines of code:
S={'A','B','C'}
U={'A','Z','C'}
U.union(S)
#What is the intersection of set S and U?
S={'A','B','C'}
U={'A','Z','C'}
print(S.intersection(U))
```
