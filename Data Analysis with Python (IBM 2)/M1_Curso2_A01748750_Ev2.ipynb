#### Getting Started Analyzing Data in Python 4:14

Una vez que ya hemos aprendido a cargar/descargar los archivos, podemos pensar en analizar sus datos, pero antes de hacerlo, primero debemos entenderlos, revisando, por ejemplo, el tipo de datos, su distribución y si tienen algún tipo de problema.
Los principales tipos de datos almacenados en pandas son:
- objetos. Equivalente a string en Python ('texto1')
- int64. Equivalente a int en Python (Números enteros: 1,2)
- float64. Equivalente a float en Python (Números decimales: 1.0,1.2)
- datetime64, timedelta[ns]. Sin equivalente directo en Python, pero se puede revisar el módulo 'datetime' en las librerías estándar.

Verificar el tipo de datos es importante por dos razones:
1. Pandas asigna los tipos automáticamente conforme la codificación que detecte de la tabla de datos original y dicha asignación puede no siempre ser correcta.
2. Asegurar la compatibilidad con otros métodos que, por ejemplo, requieran un tipo específico de datos.

Comprobando el tipo de datos contenidos en cada columna

```python
df.dtypes #Obtiene el tipo de datos asignados en cada columna
```

Obtener un resumen estadístico de cada columna para entender mejor la distribución de los datos, como si hay valores extremos o muchas dispersión.

```python
df.describe() #Descripción estadística de las columnas del dataframe
```

De forma automática .describe() omite las columnas que no contienen números. Si queremos incluirlas, usar el argumento .describe(include='all').

```python
df.describe(include='all') #Incluyendo objetos
#Se observan algunos valores NaN = Not a Number
```

Otro método para revisar el conjunto de datos es .info(), que nos muestra información de las columnas.

```python
df.info()
```

#### Lab 1:Introduction

Carga del archivo csv desde internet

```python
# Importar librerías
import pandas as pd
import numpy as np
path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
df = pd.read_csv(path, header=None)
df
```

Cambiar los títulos de las columnas (paso necesario para una pregunta)

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n", headers)
df.columns = headers
df.head(10)
```

Proceso para cambiar valores inválidos (?) a NaN y posteriomente eliminar sus filas.

```python
df1=df.replace('?',np.NaN)
df=df1.dropna(subset=["price"], axis=0)
df.head(20)
```
