#### Pre-Processing Data in Python 2:09

El  preprocesamiento de datos es un paso necesario previo al análisis de datos, es también llamado 'Limpieza de datos' (data cleaning / data wrangling), algunos pasos:
- Identificar y lidiar con valores faltantes. Un dato faltante es la condición que ocurre cuando la entrada donde debería haber un dato está vacía.
- Formato de los datos. Ya que pueden venir de distintas fuentes, también en diferentes formatos, unidades o convenciones, siendo posible estandarizarlos para se encuentren en igual situación.
- Normalización de datos (centrado/escalado). Las diferentes categorías pueden contener datos numéricos con diferentes rangos, cuya comparación podría no tener mucho significado, por lo que la normalización implica poner todos los datos dentro dle mismo rango.
- Binarización de los datos (Hacerlos binarios). Crea grandes categorias a partir de conjuntos de valores numéricos, siendo más útil la comparación.
- Convertir datos categóricos a numéricos. Esto supone un modelado estadístico más sencillo.

#### Dealing with missing values in Python

Usualmente cuando se tiene un valor faltante en una observación particular aparece como '?', 'N/A' o simplemente una calda vacía. Algunas opciones típicas para lidiar con ellos son:
- Investigar la fuente que generó los datos y preguntar qué valor debió ser
- Remover los datos donde se encuentra el valor faltante, ya sea eliminando toda la observación o solo esa entrada particular (dependiendo del tamaño que tenga nuestro conjunto de datos)
- Reemplazando. No se desperdician datos pero es menos preciso, ya que debemos reemplazar al faltante con lo que creemos podría ser (imputar), ya sea por el promedio de la variable entera (columna), la moda (valor más frecuente), con otras funciones como fundamento (si conocemos relativamente el comportamiento en función de otra variable) o simplemenento mantenerlos como datos pérdidos

*.dropna()* para remover datos con pérdidos, como los NaN. Aplicable en filas o columnas

**df.dropna()**

axis= 0 - Elimina la fila entera

axis= 1 - Elimina la columna entera

El argumento 'inplace=True' indica que haga la modificación el df directamente

```python
print(df.isnull().sum()) #Verificamos dónde hay valores faltantes
df.dropna(subset=['price'], axis=0) #NO modifica el df
df.dropna(subset=['price'], axis=0, inplace=True) #SÍ modifica el df
#Ya que el precio es lo que deseamos predecir, eliminamos las filas donde falte ese valor, 
#aunque en este caso no los había, como nos arroja la primera línea de código
```

#### Data Normalization in Python 3:34

La normalización es una técnica importante en el preprocesamiento de datos. **Al normalizar los datos, creamos un rango consistente entre las variables**, lo que nos permite hacer comparaciones confiables entre las características, debido a que nos aseguramos que tengan el mismo impacto, resultando importante también por razones computacionales.

En algunos análisis, por ejemplo en la regresión lineal, una variable con un rango mayor puede tener mayor influencia, pero no porque sea en realidad más importante como predictor, por lo que la naturaleza de los datos puede sesgar el modelo. Normalizamos para evitar dicho conflicto, por ejemplo, en un rango de 0-1. 

Algunas técnicas son:

- **Escalado de características simple (simple feature scaling).** Crea un rango entre 0 a 1

X_new= X_old / X_max 

- **Método min/máx.** Crea un rango entre 0 a 1

X_new= (X_old - X_min) / (X_max - X_min)

- **Z-score (standard score).** Crea típicamente un rango entre -3 a 3, aunque puede ser mayor

Z_new= (X_old - Mu) / sigma  , donde:

Mu= Promedio de la característica
sigma= Desviación estándar

Creamos un nuevo dataframe para probar las técnicas de normalización

```python
personas= [['Edad', 20, 30, 40, 50, 60],    #Lista para contener los valores
        ['Estatura (cm)', 150, 160, 170, 180, 190],
        ['Sueldo (mxn)', 10000, 15000, 20000, 25000, 30000],
        ['Años de estudio', 12, 15, 19, 21, 24]]
df_personas=pd.DataFrame(personas).transpose() #Convertimos la lista a un dataframe
nomCols=df_personas.iloc[0] #Extraemos la primera fila con los nombres de las columnas
df_personas.columns=nomCols #Renombramos las columnas
df_personas.drop(0, axis=0, inplace=True) #Eliminamos la fila que contenía los nombres
#Aunque el proceso anterior podría haberse hecho creando dos listas separadas, sin necesidad de
#incluir los nombres y luego eliminarlos, me pareció más fácil hacerlo así debido a que
#así tengo presentes los nombres y orden de las columnas todo el tiempo

df_personas
```

Escalado de características simple:

**X_new= X_old / X_max**

```python
df_n1=df_personas.copy()
for i in range(0,4):
  df_n1.iloc[:,i]=df_n1.iloc[:,i]/df_n1.iloc[:,i].max()
  #df_n1['Estatura (cm)']=df_n1['Estatura (cm)']/df_n1['Estatura (cm)'].max()
df_n1
```

Método min/máx:

**X_new= (X_old - X_min) / (X_max - X_min)**

```python
df_n2=df_personas.copy()
for i in range(0,4):
  df_n2.iloc[:,i]=(df_n2.iloc[:,i]-df_n2.iloc[:,i].min())/(df_n2.iloc[:,i].max()-df_n2.iloc[:,i].min())
df_n2
```

Z-score (standard score):

**Z_new= (X_old - Promedio) / DEST**


```python
df_n3=df_personas.copy()
for i in range(0,4):
  df_n3.iloc[:,i]=(df_n3.iloc[:,i]-df_n3.iloc[:,i].mean())/(df_n3.iloc[:,i].std())
df_n3
```

#### Binning in Python 1:53

El agrupamiento como preprocesamiento de datos implica crear contenedores para los valores, es decir, hacerlos categóricos, pues tal representación nos puede ayudar a entender mejor su distribución.

Creamos un dataframe con edades que van desde cero hasta 100, de 5 en 5 años, para crear categorías conforme a las etapas del ser humano

```python
edades= [range(0,100,5)] #Lista para contener los valores de edades
df_años=pd.DataFrame(edades).transpose()  #Convertimos la lista a un dataframe
df_años.columns=['Edades']
df_años
```

Creamos un arreglo para separar cinco etapas considerar en la vida del ser humano, al igual que una lista con los nombres para usar la función **pd.cut()** y agrupar

Nota: Esto tiene un comportamiento meramente didáctico, pues las edades suelen ser consideradas más como valores discretos y el espacio entre una etapa y otra no es uniforme, pero permite ejemplificar el agrupamiento

```python
import random
bins=np.linspace(min(df_años['Edades']),max(df_años['Edades']),6)
etapas=['Infancia','Adolescencia','Juventud','Adultez','Persona Mayor']
df_años['Edades-bin']=pd.cut(df_años['Edades'], bins, labels=etapas, include_lowest=True)
df_años
```

Creamos un gráfico de barrar para visualziar mejor las categorías, mas como el arreglo de valores fue creado con separación uniforme, todas tendrán la misma cantidad de valores

```python
import matplotlib as plt
from matplotlib import pyplot
#plt.pyplot.hist(x=edades,y=df_años)
pyplot.bar(etapas, df_años['Edades-bin'].value_counts())

plt.pyplot.xlabel("Edades-bin")
plt.pyplot.ylabel("Valores")
plt.pyplot.title("Edades bin")
```
